```python
cha_count = {}
def append_cha_count(txt):
    for cha in list(set(txt)):
        cha_count[cha] = cha_count.get(cha, 0) + 1
        
train['review'].map(append_cha_count)

print(cha_count)
```

```python
df_char = pd.DataFrame()
df_char['char'] = list(cha_count.keys())
df_char['count'] = list(cha_count.values())
df_char = df_char.sort_values('count', ascending = False).reset_index(drop = True)
```

```python
df_char.loc[df_char['count'] >= 10].sample(20)
df_char = df_char.loc[df_char['count'] >= 10]
```

```python
exclude_chars = ['\n', ' ', '\\', '¬∞', '.', "'", '"', '-', '!', '-', ')', '(', ':', ';',
                ',', '\u200b', 'Ô∏è', '/', '\xa0', "?", "&", '‚Äô', '~', '+', '*', '%', '_', 
                 '#', '=', '@', ']', '[']

mask = df_char['char'].map(lambda x: not (x.isalnum() or str(x) in exclude_chars))
print(mask.head(10))
```

```python
df_char_special = df_char.loc[mask]
print(df_char_special)
```

```python
def counting_char_with_sentiment(char, sentiment):
    return sum([char in review for review in list(train.loc[train['rating'] == sentiment]['review'])])
```

```python
for i in range(1,6):
    df_char_special[f'rating_{i}'] = np.vectorize(counting_char_with_sentiment)(df_char_special['char'], i)
    
```

```python
df_char_special.head(10).style.background_gradient(cmap = 'Greens')
```

```python
rating_cols = [f'rating_{i}' for i in range(1,6)]
weighted_score = list(np.arange(-1,1.5,0.5))

print(rating_cols)
df_char_special['score'] = np.array([list(df_char_special[a] * b) for a,b in zip(rating_cols, weighted_score)]).sum(axis = 0) / df_char_special[rating_cols].sum(axis = 1)
df_char_special.sort_values(by = 'score', ascending = False).head(10)[['char','score'] + rating_cols].style.background_gradient(cmap = 'Greens')
```
