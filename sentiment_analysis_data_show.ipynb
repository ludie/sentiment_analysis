```python
def ngrams_top(corpus,ngram_range,n=None):

    vec = CountVectorizer(stop_words = 'english',ngram_range=ngram_range, min_df= 1).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    total_list=words_freq[:n]
    df=pd.DataFrame(total_list,columns=['text','count'])
    return df

def merge_sentiment_ngrams(df, ngram_range,n=None):
    df_full = ngrams_top(df['review'],ngram_range,n)
    
    for i in range(1,6):
        df_temp = ngrams_top(df.loc[df['rating'] == i,'review'],ngram_range,n)
        df_temp.columns = ['text', f'rating_{i}']
        
        df_full = df_full.merge(df_temp, on = 'text', how = 'outer')
        df_full = df_full.fillna(0)
        df_full[f'rating_{i}'] = df_full[f'rating_{i}'].astype(int)
        
    
    df_full = df_full.loc[df_full['count'] >= 3 ]
    
    rating_cols = [f'rating_{i}' for i in range(1,6)]
    weighted_score = list(np.arange(-1,1.5, 0.5))

    df_full['score'] = np.array([list(df_full[a] * b) for a,b in zip(rating_cols, weighted_score)]).sum(axis = 0) /  df_full[rating_cols].sum(axis = 1)
    return df_full
```

```python
df_1grams = merge_sentiment_ngrams(train, ngram_range = (1, 1))
df_2grams = merge_sentiment_ngrams(train, ngram_range = (1, 2))

df_1grams = df_1grams.loc[df_1grams['count'] >= 100].sort_values('score', ascending = False)
df_2grams = df_2grams.loc[df_2grams['count'] >= 100].sort_values('score', ascending = False)
```
